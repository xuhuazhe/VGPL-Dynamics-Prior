```python
def plot_curves(iters, loss):
    plt.figure(figsize=[16,9])
    plt.plot(iters, loss)
    plt.xlabel('iterations', fontsize=30)
    plt.ylabel('emd loss', fontsize=30)
    plt.title('Training Loss', fontsize=35)
    plt.xticks(fontsize=25)
    plt.yticks(fontsize=25)
#     plt.plot(loss)
    
```

```python
with open('../dump/dump_Pinch/files_dy26-Aug-2021-17:37:34.343845_nHis4_aug0.05emd_seqlen20/train.npy', 'rb') as f:
    train_log = np.load(f, allow_pickle=True)
    train_log = train_log[None][0]
    
```
